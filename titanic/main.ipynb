# Juntando os datasets de treino e teste

Nessa análise foi escolhido juntar os 2 datasets train e test em um só para depois separa-los novamente. Essa opção garante que toda feature engineering (processo de criar variaveis) seja usada para o dataframe como um todo, utilizando-se da mesma arquitetura.  Um modelo de machine learning irá aprender a partir dos dados que são fornecidos,esse processo se torna um dos passos mais cruciais na criação de um modelo. Construir variáveis relevantes pode ser a diferença entre ter um modelo com um ótimo desempenho ou não.

```python
# salvar os índices dos datasets para recuperação posterior
train_idx = train.shape[0]
test_idx = test.shape[0]

# salvar PassengerId para submissao ao Kaggle
passengerId = test['PassengerId']

# extrair coluna 'Survived' e excluir ela do dataset treino
target = train.Survived.copy()
train.drop(['Survived'], axis=1, inplace=True)

# concatenar treino e teste em um único DataFrame
df_merged = pd.concat(objs=[train, test], axis=0).reset_index(drop=True)

print("df_merged.shape: ({} x {})".format(df_merged.shape[0], df_merged.shape[1]))
```

```python
df_merged
```

## Seleção de Colunas/features:

Realizando a exclusão das *features* que não tem relevancia no treinamento do modelo:

Variaveis desconsideradas: 'PassengerId', 'Name', 'Ticket', 'Cabin', pois aparentemente não parecem relevantes para o proposito de prever sobrevivencia no Titatic no modelo.

```python
df_merged.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True )
```

```python
df_merged
```

## Resolvendo os valores faltantes:

Sempre que existirem valores faltantes em nosso dataset, precisamos escolher entre preencher ou excluir a linha que consta esse dado. A exclusão irá comprometer toda a linha de informação, sendo assim, o cientista de dados pode escolher a opção de preencher os dados com algum valor, seja a média, valor de maior frequencia etc e assim garantir mais um dado.  Nesse dataset vou preencher com os valores da média.

```python
# identificando a quantidade de valores faltantes ou nulos datasets de treino e teste
df_merged.isnull().sum()
```

```python
# preenchendo(fillna) a variavel age e tarifa com a média dos valores do ds
age_median = df_merged['Age'].median()
df_merged['Age'].fillna(age_median, inplace=True)

# fare
fare_median = df_merged['Fare'].median()
df_merged['Fare'].fillna(fare_median, inplace=True)

# A embarked será preenchida com o valor que mais apareceu para essa variavel
embarked_top = df_merged['Embarked'].value_counts()[0]
df_merged['Embarked'].fillna(embarked_top, inplace=True)
```

```python
#Substituindo caracteres(valores categoria) por numeros. O modelo somente interpreta os valores numericos.
df_merged['Sex'] = df_merged['Sex'].map({'male': 0, 'female': 1 })
```

```python
#utilando dummy variables para a feature Enbarked, essas variaveis assumem 0 e 1 para cada valor da variavel, criando assim colunas e marcando os valores nessas colunas para cada dado. 

# dummie variables para 'Embaked'
embarked_dummies = pd.get_dummies(df_merged['Embarked'], prefix='Embarked')
df_merged = pd.concat([df_merged, embarked_dummies], axis=1)
df_merged.drop('Embarked', axis=1, inplace=True)
```

```python
df_merged.head()
```

```python
# recuperar datasets de treino e teste utilizando a função iloc para separar em train os dados até train_idx (indices do primeiro dataset train) e de train_idx até o final dos dados para test.
train = df_merged.iloc[:train_idx]
test = df_merged.iloc[train_idx:]
```

## Construindo o Modelo

Após a análise exploratória e tratamento dos dados, etapas que demoram mais no projeto, porém são de grande importancia para se criar um modelo otimizado e poder fazer as previsões corretas.

Serão utilizados 2 algoritmos nesse modelo:

* *Modelo de Regressão Logística*: quando a variável de interesse (resposta) é binária, isto é, “sim” ou “não”. Através da Regressão Logística é possível avaliar os fatores de influenciam a ocorrência de determinado evento. No nosso caso a variável resposta é a morte ou a sobrevivência do passageiro ('Survived').


* *Modelo de Árvore de Decisão*: ideia de divisão dos dados em grupos homogêneos para as decisões, que serão tomadas a partir de uma amostra de teste. Nela temos os nós folhas onde é atribuído a classe (Yes/No) no caso da classificação e a média das observaçoes no caso da regressão.


```python
# importar bibliotecas do modelo de Machine Learning
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(train, target)

# verificar a acurácia do modelo
acc_logReg = round(lr_model.score(train, target) * 100, 2)
print("Acurácia do modelo de Regressão Logística: {}".format(acc_logReg))
```

```python
#treinando com o modelo  de regressao para os dados de test e criando o arquivo para submeter ao Keggle. 
y_pred_lr = lr_model.predict(test)

submission = pd.DataFrame({
    "PassengerId": passengerId,
    "Survived": y_pred_lr
})

# gerar arquivo csv
submission.to_csv('./submission_lr.csv', index=False)
```

```python
#Utilizando o modelo de arvore de decisão para treinar os dados train e verificar se a acuracia é melhor do que o modelo de regressao testado anteriormente.

#criando o modelo na variavel tree_model ou a maquina preditiva
tree_model = DecisionTreeClassifier(max_depth=3)
tree_model.fit(train, target)

# verificar a acurácia do modelo
acc_tree = round(tree_model.score(train, target) * 100, 2)
print("Acurácia do modelo de Árvore de Decisão: {}".format(acc_tree))
```

```python
#criando outro arquivo para submeter o keggle com a previsao do algoritmo de arvore
y_pred_tree = tree_model.predict(test)

submission = pd.DataFrame({
    "PassengerId": passengerId,
    "Survived": y_pred_tree
})

# gerar arquivo csv
submission.to_csv('./submission_tree.csv', index=False)
```

```python
#Comando de Salvamento da Máquina Preditiva
import pickle 
pickle_out = open("maquina_preditiva.pkl", mode = "wb") 
pickle.dump(tree_model, pickle_out) 
pickle_out.close()
```
